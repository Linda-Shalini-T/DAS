```python
import pandas as pd

# Load the dataset to examine its structure and contents
file_path = 'Women Dresses Reviews Dataset .csv'
data = pd.read_csv(file_path)

# Display the first few rows and the columns of the dataset
data.head(), data.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a consistent style for plots
sns.set(style="whitegrid")

# Chart 1: Age distribution of reviewers
plt.figure(figsize=(6, 4))
sns.histplot(data['age'], bins=15, kde=True, color='skyblue')
plt.title('Age Distribution of Reviewers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()


# Chart 2: Average rating by department
plt.figure(figsize=(6, 4))
avg_rating_by_dept = data.groupby('department_name')['rating'].mean().sort_values()
sns.barplot(x=avg_rating_by_dept, y=avg_rating_by_dept.index, palette='coolwarm')
plt.title('Average Rating by Department')
plt.xlabel('Average Rating')
plt.ylabel('Department')
plt.tight_layout()


# Chart 3: Recommendation trends based on rating
plt.figure(figsize=(6, 4))
recommend_trend = data.groupby('rating')['recommend_index '].mean()
sns.lineplot(x=recommend_trend.index, y=recommend_trend.values, marker='o', color='green')
plt.title('Recommendation Trends by Rating')
plt.xlabel('Rating')
plt.ylabel('Recommendation Percentage')
plt.xticks(range(1, 6))
plt.tight_layout()



```
