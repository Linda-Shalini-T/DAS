```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    "Time": pd.date_range("2024-12-01", periods=10, freq="h"),
    "PM2.5": [35, 37, 34, 36, 33, 31, 32, 30, 29, 28]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot PM2.5 levels
plt.plot(df["Time"], df["PM2.5"], marker="o", label="PM2.5")
plt.xlabel("Time")
plt.ylabel("PM2.5 Level")
plt.title("Air Quality Index")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
