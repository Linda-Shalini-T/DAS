```python
import pandas as pd

# Load the uploaded CSV file to understand its structure
file_path = 'skin_care.csv'
skin_care_data = pd.read_csv(file_path)

# Display the first few rows of the dataset and its summary to understand its structure and contents
skin_care_data.head(), skin_care_data.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Clean the dataset by removing unnecessary columns
skin_care_data = skin_care_data.drop(columns=['Unnamed: 0'])

# Visualization 1: Price vs. Star Ratings
plt.figure(figsize=(10, 6))
sns.scatterplot(data=skin_care_data, x='price', y='stars', size='vote', hue='stars', palette='cool', sizes=(20, 200))
plt.title('Price vs. Star Ratings of Skincare Products', fontsize=16)
plt.xlabel('Price ()', fontsize=12)
plt.ylabel('Star Ratings', fontsize=12)
plt.legend(title='Star Ratings', loc='upper left')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Visualization 2: Top-rated products based on number of votes
top_rated = skin_care_data.nlargest(10, 'stars').sort_values('vote', ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(data=top_rated, x='vote', y='title', hue='stars', dodge=False, palette='viridis')
plt.title('Top 10 Rated Skincare Products', fontsize=16)
plt.xlabel('Number of Votes', fontsize=12)
plt.ylabel('Product Title', fontsize=12)
plt.legend(title='Star Ratings')
plt.tight_layout()
plt.show()
```
