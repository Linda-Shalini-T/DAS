```python
import pandas as pd
# Load the BMW pricing challenge dataset to examine its structure and contents
bmw_data_path = 'bmw_pricing_challenge.csv'
bmw_data = pd.read_csv(bmw_data_path)

# Display the first few rows and the columns of the dataset
bmw_data.head(), bmw_data.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Chart 1: Mileage distribution
plt.figure(figsize=(6, 4))
sns.histplot(bmw_data['mileage'], bins=20, kde=True, color='skyblue')
plt.title('Mileage Distribution of BMW Cars')
plt.xlabel('Mileage (km)')
plt.ylabel('Frequency')
plt.tight_layout()


# Chart 2: Average price by fuel type
plt.figure(figsize=(6, 4))
avg_price_by_fuel = bmw_data.groupby('fuel')['price'].mean().sort_values()
sns.barplot(x=avg_price_by_fuel.index, y=avg_price_by_fuel.values, palette='muted')
plt.title('Average Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price (€)')
plt.tight_layout()


# Chart 3: Correlation between engine power and price
plt.figure(figsize=(6, 4))
sns.scatterplot(data=bmw_data, x='engine_power', y='price', hue='fuel', palette='cool', alpha=0.7)
plt.title('Engine Power vs. Price')
plt.xlabel('Engine Power (kW)')
plt.ylabel('Price (€)')
plt.legend(title='Fuel Type', bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()


```
